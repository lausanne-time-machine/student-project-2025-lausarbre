```python
csv_filenames = os.listdir("data/csv")
```

```python
num_records = {}
counts = {}
for filename in csv_filenames:
    df = pd.read_csv(f"data/csv/{filename}")
    num_records[filename] = df.shape[0]
    for col in df.columns:
        if not col in counts:
            counts[col] = 1
        else:
            counts[col] += 1
```
